# Youtube Top 1000 Most Subscriptd Channel Analysis

## Read the Dataset

```r
df <- read.csv ("/content/most_subscribed_youtube_channels.csv")
head (df)
```

## Dataset  Summary

```r
summary (df)
```

### Pre-Process Data

###Remove commas from the numbers. 

```r
df$video.count <- as.numeric(gsub(",", "", df$video.count))
df$video.views <- as.numeric(gsub(",", "", df$video.views))
df$subscribers <- as.numeric(gsub(",", "", df$subscribers))

head (df)
```

### Remove the Channel with zero video.count since they have to marketing value. 

```r
library (dplyr)
df<-df%>% filter (video.count != 0)
head (df)
```

### Check for Unfilled rows in category Column

```r
df[df$category == "", ]
```

###Remove the unfilled categories

```r
df<-filter(df, category != "")
head (df)
```

####Format started column

Started by themselves have no statistical value but by converting it to number of years it as existed we can ooze out it in data-processing. 

```r
df$started <- 2022 - as.numeric(df$started)
head (df)
```

## Exploratory Data Analysis

### Exploring the category column.

#### Distinct Categories in the column.

```r
n_distinct (df$category)
```

#### Frequency Dataframe for categories

```r
require(tibble)
```

```r
freq_category <- data.frame(enframe(apply(data.frame(df$category), 2, table), name="Categories", value="Frequency"))
freq_category <- freq_category[order(freq_category$Frequency, decreasing = TRUE), ]
freq_category
```

 #### Top 8 Popular Categories in Youtube


```r
require(ggplot2)
```

```r
data<-top_n (freq_category, 8, Frequency)
data
```

```r
ggplot(data=data, aes(x=Categories, y=Frequency)) +
  geom_bar(stat="identity", fill="steelblue")+
  theme_minimal()
```

#### Top 8 least popular categories in Youtube

```r
data<-top_n (freq_category, -8, Frequency)
data
```

```r
ggplot(data=data, aes(x=Categories, y=Frequency)) +
  geom_bar(stat="identity", fill="chartreuse4")+
  theme_minimal()
```

####Parliament Plot for category

```r
install.packages("ggparliament")
```

```r
require(ggparliament)
```

```r
circle <- parliament_data(election_data = freq_category,
                             type = "circle",
                             parl_rows = 10,
                             party_seats = freq_category$Frequency)
```

### Exploring Subscribers column

#### Top 7 channels with most amount of subscribers.

```r
top_7_subs <- top_n(df, 7, subscribers)
top_7_subs
```

```r
ggplot(data=top_7_subs, aes(x=Youtuber, y=subscribers)) +
  geom_bar(stat="identity", fill="steelblue")+
  theme_minimal()
```

#### Top 7 channels with least amount of subscribers.

```r
least_7_subs <- top_n(df, -7, subscribers)
least_7_subs
```

```r
ggplot(data=least_7_subs, aes(x=Youtuber, y=subscribers)) +
  geom_bar(stat="identity", fill="chartreuse4")+
  theme_minimal()
```

#### Plot Subscriber count vs View count by grouping them wrt to Category

```r
ggplot (df, aes(subscribers, video.views)) + geom_point (aes(col=category)) + geom_smooth(method=lm, se=TRUE, color="yellow")
```

#### Box Plot for measuring the outliers. 

```r
ggplot (df, aes(subscribers)) + geom_boxplot(aes(y=subscribers))
```

#### Density Plot


```r
density_followers <- ggplot(df, aes(x=subscribers)) + geom_density ()
density_followers + geom_vline(aes(xintercept=mean(subscribers)), color="red", linetype="dashed", linewidth=1) + geom_vline(aes(xintercept=median(subscribers)), color="blue", linetype="dashed", linewidth=1)
```

#### Test the Correlation Hypothesis with view count.

```r
test<-cor.test (df$video.views, df$subscribers)
test
```

Null Hypothesis failed, implies number of subscribers are related to video.views
